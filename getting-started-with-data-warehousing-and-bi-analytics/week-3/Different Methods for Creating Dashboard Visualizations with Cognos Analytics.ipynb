# Getting Started with Cognos Analytics

### Access to the IBM Cognos Analytics

* Download the dataset

```sh
wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0130EN-SkillsNetwork/Hands-on%20Labs/Lab%204%20-%20Getting%20Started%20with%20Cognos%20Analytics/CustomerLoyaltyProgram.csv
```

Upload the dataset to `Cognos` from the `Upload data` option in the dashboard. It can be checked by the `Content` option in the navigation bar.

The Template window will display allowing to select the type of dashboard and the template style. Select the **tabbed dashboard style**. This will allow you to have multiple pages for your dashboards. Select the **three-panel template**. Click **Create**.

* `First Dashboard` in `Cognos` and exploring the options
